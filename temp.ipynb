```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
t=np.arange(0.0,2.0,0.1)
s=np.sin(t*np.pi)
plt.subplot(2,3,1) #要生成两行两列，这是第一个图plt.subplot('行','列','编号')
plt.plot(t,s,'b--')
plt.ylabel('y1')
plt.subplot(2,3,3) #两行两列,这是第二个图
plt.plot(2*t,s,'r--')
plt.ylabel('y2')
plt.subplot(2,3,4)#两行两列,这是第三个图
plt.plot(3*t,s,'m--')
plt.subplot(2,3,6)#两行两列,这是第四个图
plt.plot(4*t,s,'k--')
plt.show()
```

```python
t=np.arange(0.0,2.0,0.1)
s=np.sin(t*np.pi)
figure,ax=plt.subplots(2,2)
ax[0][0].plot(t,s,'r*')
ax[0][0].plot(t*2,s,'b-*')
ax[0][0].plot(t*3,s,'y*')
ax[0][0].plot(t*4,s,'g-*')

#figure   # 全部再画一遍
ax[0][1].plot(t*2,s,'b-*')
ax[1][0].plot(t*3,s,'y*')
ax[1][1].plot(t*4,s,'g-*')
```
